```python
#Outliers : Z Score

outliers = []
def detect_outliers(data):
    threshold =3
    mean = np.mean(data)
    std = np.std(data)
    
    
    for i in data:
        z_score = (i - mean)/std
        if np.abs(z_score) > threshold:
            outliers.append(i)
    return outliers
```

```python
outlier_pss = detect_outliers(base.pss)
print(len(outlier_pss)) #1
```

```python
#converting outlier values to missing values

def outlier_limits(col):
    Q3, Q1 = np.nanpercentile(col, [75,25])
    IQR = Q3 - Q1
    UL = Q3 + 1.5 * IQR
    LL = Q1 - 1.5 * IQR
    return UL, LL
```

```python
for column in base.columns:
    if base[column].dtype != 'object':
        UL, LL = outlier_limits(base[column])
        base[column] = np.where((base[column] > UL) | (base[column] < LL), np.nan, base[column])
```
